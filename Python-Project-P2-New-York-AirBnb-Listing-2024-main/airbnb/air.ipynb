**BI Variable analysis**

```python
# it means one variable depends on other variable
df.columns
```

```python
# price depends on neighbourhood_group and room_type
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type', palette='Set2')
plt.title('Average Price by Neighbourhood & Room Type', fontsize=14, fontweight='bold')
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.legend(title='Room Type', fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Number of reviews and price 
plt.figure(figsize=(12,7))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group', palette='husl', s=80, alpha=0.7)
plt.title('Guest Reviews vs Price - Market Analysis by Neighbourhood', fontsize=14, fontweight='bold')
plt.xlabel('Number of Reviews', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.legend(title='Neighbourhood', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=df, vars=['price','minimum_nights','number_of_reviews','availability_365'] , hue='room_type')
```

```python
#Geographical Distribution of airbnb Listings
plt.figure(figsize=(14,8))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type', palette='Set1', s=100, alpha=0.6)
plt.title('Geographical Distribution of Airbnb Listings Across NYC', fontsize=14, fontweight='bold')
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend(title='Room Type', fontsize=10, loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Heat map - correlation of one variable with other for numerical columns
corr = df[['latitude', 'longitude', 'price', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
plt.figure(figsize=(12, 9))
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0, square=True, linewidths=1, cbar_kws={'label': 'Correlation'}, fmt='.2f')
plt.title('Correlation Heatmap - Variable Relationships Analysis', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
